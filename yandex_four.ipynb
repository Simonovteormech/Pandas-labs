```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv(r"C:\Users\Leher\Downloads\PokemonData.csv")
df
```

1) Постройте гистограмму распределения показателя здоровья (HP) у покемонов

```python
plt.hist(df['HP'], label = 'HP')
plt.xlabel("Health point")
plt.ylabel("Amount")
plt.title('Hist №1')
plt.legend()
plt.show()
```

```python
df['HP'].max()
```

2) Гистограмма: две переменные на одном графике

```python
plt.hist(df['HP'], label = 'HP')
plt.hist(df['Attack'], label = 'Attack')
plt.xlabel("HP/Attack")
plt.ylabel("Amount")
plt.title('Hist №2')
plt.legend()
plt.show()
```

3) Гистограмма: прозрачность

```python
plt.hist(df['HP'], label = 'HP', alpha=0.5)
plt.hist(df['Attack'], label = 'Attack', alpha=0.5)
plt.xlabel("HP/Attack")
plt.ylabel("Amount")
plt.title('Hist №2')
plt.legend()
plt.show()
```

4) Точечная диаграмма

```python
plt.scatter(df['Attack'], df['Defense'], alpha=0.3)
plt.show()
```

5) Столбчатая диаграмма: несколько переменных

```python
df.groupby('Legendary')['Type1'].value_counts().unstack(0).plot(kind = 'bar')
plt.show()
```

6) Столбчатая диаграмма

```python
df['Type1'].value_counts().plot(kind = 'bar', xlabel='Тип покемонов', ylabel='Количество', title='Легендарные покемоны по типам в сравнении с обычными')
plt.show()
```
